```python
model_path = "C:/Users/lamtr/Documents/my_model_waste.h5"
```

```python
from PIL import Image
from keras import models
model = models.load_model(model_path, custom_objects=None, compile=True, options=None)
video = cv2.VideoCapture(0)
text_color = (206,235,135)
org = (50,50)
font = cv2.FONT_HERSHEY_SIMPLEX
fontScale = 1
thickness = 3
class_list = ['Recycle','Organic','No']

def advice (predict_1):
    if predict_1 == 0:
        print('You need to recycle it')
    elif predict_1 == 1:
        print('No recycling!!!')
    else :
        print('Please throw it into the trash')
while True:
    ret, image_np = video.read()
    image_np = cv2.flip(image_np, 1)
    image_resized = cv2.resize(image_np, (80, 80))
    img_array = tf.expand_dims(image_resized, 0)
    predict = model.predict(img_array)
    predict_ = np.where(predict >=0.5, 1,0)
    predict_1 = int(predict_ >= 0.5)
    print(predict_1)
    predicted_class = class_list[predict_1]
    image_np = cv2.putText(image_np, "Image Classification Output: " + str(predicted_class), org,
      font, fontScale, text_color, thickness, cv2.LINE_AA)
    print(predict)
        
    cv2.imshow('Capturing', image_np)
    advice(predict_1)
    key = cv2.waitKey(1)
    if key == ord('q'):
        break

video.release()
cv2.destroyAllWindows()
```
